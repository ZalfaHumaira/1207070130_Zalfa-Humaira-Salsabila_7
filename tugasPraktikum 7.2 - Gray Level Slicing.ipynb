```python
#import library
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('tootles.jpg', cv2.IMREAD_GRAYSCALE) #membaca gambar dari file yang disimpan menggunakan opencv

row, column = img.shape #digunakan untuk mendapatkan dimensi atau ukuran citra 

img1 = np.zeros((row,column),dtype = 'uint8')#membuat array dengan ukuran yang sama
 
#menentukan nilai minimun dan maksimum kontras
min_range = 10
max_range = 60

#fungsi perulangan yang digunakan untuk menentukan filter batas
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8)) #membuat tata letak subplot
ax = axes.ravel() #digunakan untuk meratakan array 

ax[0].imshow(img, cmap=plt.cm.gray) #menampilkan gambar
ax[0].set_title("Citra Input") #menambahkan judul pada gambar
ax[1].hist(img.ravel(), bins=256) #menampilkan histogram
ax[1].set_title('Histogram Input') #menambahkan judul pada histogram

ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
