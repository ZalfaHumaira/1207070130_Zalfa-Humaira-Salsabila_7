```python
#import library
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
import cv2
```

```python
citra1 = cv2.imread("laba.png", cv2.IMREAD_GRAYSCALE) #membaca gambar yang tersimpan lalu mengubahnya ke format grayscale
citra2 = cv2.imread("miaw.png", cv2.IMREAD_GRAYSCALE) #membaca gambar yang tersimpan lalu mengubahnya ke format grayscale

print('Shape citra 1 : ', citra1.shape) #menampilkan shape gambar laba
print('Shape citra 2 : ', citra2.shape) #menampilkan shape gambar miaw

fig, axes = plt.subplots(1, 2, figsize=(10, 10)) #membuat tata letak subplot
ax = axes.ravel() #digunakan untuk meratakan array 

ax[0].imshow(citra1, cmap = 'gray') #menampilkan gambar laba
ax[0].set_title("Citra 1") #set judul gambar
ax[1].imshow(citra2, cmap = 'gray') #menampilkan gambar miaw
ax[1].set_title("Citra 2") #set judul gambar 
```

```python
copyCitra1 = citra1.copy().astype(float) #copy gambar pada citra1 dan menyimpan dalam tipe data float
copyCitra2 = citra2.copy().astype(float) #copy gambar pada citra2 dan menyimpan dalam tipe data float
#mendapatkan ukuran gambar
m1,n1 = copyCitra1.shape 
output1 = np.empty([m1, n1])
m2,n2 = copyCitra2.shape
output2 = np.empty([m2, n2])

print('Shape copy citra 1 : ', copyCitra1.shape) #menampilkan shape gambar laba 
print('Shape output citra 1 : ', output1.shape) #menampilkan shape gambar laba yang di copy ke tipe data float

print('m1 : ',m1)
print('n1 : ',n1)
print()

print('Shape copy citra 2 : ', copyCitra2.shape)#menampilkan shape gambar miaw
print('Shape output citra 3 : ', output2.shape)#menampilkan shape gambar miaw yang di copy ke tipe data float
print('m2 : ',m2)
print('n2 : ',n2)
print()
```

```python
#fungsi yang digunakan untuk menerapkan filter rerata pada citra 1
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        a1 = baris
        b1 = kolom
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1];  
        output1[a1, b1] = (1/9 * jumlah) #mengalikan setiap elemen citra dengan 1.9
```

```python
#fungsi yang digunakan untuk menerapkan filter rerata pada citra 2
for baris1 in range(0, m2-1):
    for kolom1 in range(0, n2-1):
        a1 = baris1
        b1 = kolom1
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1-1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1];  
        output2[a1, b1] = (1/9 * jumlah) #mengalikan setiap elemen citra dengan 1.9
```

```python
fig, axes = plt.subplots(2, 2, figsize=(11, 11)) #membuat tata letak subplot
plt.subplots_adjust(wspace=0.2, hspace=-0.3) #menentukan besarnya jarak gambar
ax = axes.ravel() #digunakan untuk meratakan array

ax[0].imshow(citra1, cmap = 'gray') #menampilkan gambar
ax[0].set_title("Input Citra 1")#menambahkan judul pada gambar

ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Input Citra 1")

ax[2].imshow(output1, cmap = 'gray')
ax[2].set_title("Output Citra 1")

ax[3].imshow(output2, cmap = 'gray')
ax[3].set_title("Output Citra 2")
```
