```python
#import library 
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np

import cv2
```

```python
#membaca gambar yang disimpan pada file
citra1 = imread(fname="tootles.jpg")
print(citra1.shape)#menampilkan shape pada gambar yang dibaca

plt.imshow(citra1, cmap='gray')#menampilkan gambar
```

```python
#melakukan proses konvolusi untuk membuat kernel dengan matriks 3 x 3
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])
#menerapkan proses konvolusi 
citraOutput = cv2.filter2D(citra1, -1, kernel)

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) #membuat tata letak subplot
ax = axes.ravel()#digunakan untuk meratakan array

ax[0].imshow(citra1, cmap = 'gray') #menampilkan gambar original
ax[0].set_title("Citra Input") #menambahkan judul pada gambar
ax[1].imshow(citraOutput, cmap = 'gray') #menampilkan gambar hasil konvolusi
ax[1].set_title("Citra Output") #menambahkan judul pada gambar
```
